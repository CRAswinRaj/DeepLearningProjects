# Using Deep Learning in Mnist dataset 
Mnist dataset contains pictures of digits. Here, I used deep learning with the help of tensorflow library to create a model that could predict the number on the picture. 

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
```

## Collecting Data

```python
(train_data, train_labels), (test_data, test_labels) = mnist.load_data()
```

## Creating Deep Learning models

### Model 1

```python
tf.random.set_seed(42)

model1 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(5, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model1.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])

history1 = model1.fit(train_data, train_labels, epochs=10, validation_data=(test_data, test_labels))
```

```python
import pandas as pd
pd.DataFrame(history1.history).plot()
```

Accuracy in the test dataset using model 1 = 66.28% 

### Model 2

Increasing number of neurons in the third layer

```python
tf.random.set_seed(42)

model2 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model2.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])

history2 = model2.fit(train_data, train_labels, epochs=10, validation_data=(test_data, test_labels))
```

Accuracy in the test dataset using model 2 = 62.04% 

This is less than that of model 1. So increasing neurons in the third layer did not give better results

### Model 3
Increasing number of neurons in the second layer

```python
tf.random.set_seed(42)

model3 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(20, activation='relu'),
    tf.keras.layers.Dense(5, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model3.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])

history3 = model3.fit(train_data, train_labels, epochs=10, validation_data=(test_data, test_labels))
```

```python
pd.DataFrame(history3.history).plot()
```

Using model 3 we got validation accuracy = 91.90% 

This is much greater than that of model 1 and 2. Hence this change gave better result

### Model 9
Further increase neurons in 3rd layer

```python
tf.random.set_seed(42)

model9 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model9.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])

history9 = model9.fit(train_data, train_labels, epochs=10, validation_data=(test_data, test_labels))
```

```python
pd.DataFrame(history9.history).plot();
```

```python
import random
import numpy as np
```

```python
# To visualize the model, random data is taken and prediction vs real is shown
plt.figure(figsize=(9, 9))
for i in range(4):
  plt.subplot(2, 2, i+1)
  n = random.randint(0, 9999)
  y_pred = model9.predict(test_data[n].reshape((1, 28, 28)))
  predicted = y_pred.argmax()
  real = test_labels[n]
  if real == predicted:
      clr='g'
  else:
      clr='r'
  plt.imshow(test_data[n])
  plt.title(f'True label: {real} - Predicted: {predicted}', fontdict={'color':clr});
```

### Normalizing the data
The values in the data is converted to values between 0 and 1.

This is done dividing all values by 255(maximum value)

```python
import numpy as np
np.array([test_data[0]]).shape
```

```python
train_data_norm = train_data/255
test_data_norm = test_data/255
```

### Model 10

The normalised data is used for training the data. All the hyperparamers remain same as that of model 9.

```python
tf.random.set_seed(42)

model10 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model10.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])

history10 = model10.fit(train_data_norm, train_labels, epochs=10, verbose=0)
```

```python
model10.evaluate(test_data_norm, test_labels)
```

```python
pd.DataFrame(history10.history).plot();
```

After normalizing, the model performs much better from the first epoch itself. And the final validation accuracy is also significantly higher than before.

### Model 11

Changing learning rate

```python
tf.random.set_seed(42)

model11 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model11.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              optimizer=tf.keras.optimizers.Adam(learning_rate=8e-4),
              metrics=['accuracy'])

history11 = model11.fit(train_data_norm, train_labels, epochs=10, validation_data=(test_data_norm, test_labels))
```

```python
pd.DataFrame(history11.history).plot()
```

```python
model11.evaluate(test_data_norm, test_labels)
```

The model doesn't give better results after trying various learning rates

Thus, from all the 11 models, model 10 seems to be the best one.
